```python
import pandas as pd
df = pd.read_csv("/train.csv")
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# ==========================================
# FAZ 2: VERİ TEMİZLEME & ÖZELLİK ÜRETME (Belge Gün 8)
# ==========================================

# 1. Çok fazla eksik veri içeren 'Cabin' sütununu kaldırıyoruz
df_clean = df.drop(columns=['Cabin'])

# 2. 'Age' (Yaş) sütunundaki boşlukları yolcuların ortalama yaşıyla dolduruyoruz
ortalama_yas = df_clean['Age'].mean()
df_clean['Age'] = df_clean['Age'].fillna(ortalama_yas)

# 3. Belgede istenen "Yeni Özellik" (Aile Büyüklüğü) oluşturma
df_clean['Family_Size'] = df_clean['SibSp'] + df_clean['Parch']


# ==========================================
# FAZ 2 & 3: ANALİZLER VE GRAFİKLER (Belge Gün 11-14)
# ==========================================

print("=== CRITICAL INSIGHTS (Kritik Bulgular) ===")

# Analiz 1: Cinsiyete Göre Hayatta Kalma Oranı
print("\n1. Cinsiyete Göre Hayatta Kalma Oranları (%):")
print(df_clean.groupby('Sex')['Survived'].mean() * 100)

# Analiz 2: Bilet Sınıfına Göre Hayatta Kalma Oranı
print("\n2. Bilet Sınıfına (Pclass) Göre Hayatta Kalma Oranları (%):")
print(df_clean.groupby('Pclass')['Survived'].mean() * 100)

# Analiz 3: Yolcuların Genel Demografik Özetleri
print(f"\nYolcuların Ortalama Yaşı: {df_clean['Age'].mean():.2f}")
print(f"En Yaşlı Yolcu: {df_clean['Age'].max()} | En Genç Yolcu: {df_clean['Age'].min()}")

# --- GRAFİKLER ---
plt.figure(figsize=(12, 5))

# Grafik 1: Cinsiyete Göre Hayatta Kalma Sayıları
plt.subplot(1, 2, 1)
sns.countplot(x='Survived', hue='Sex', data=df_clean, palette='Set2')
plt.title('Cinsiyete Göre Hayatta Kalma (0=Öldü, 1=Kurtuldu)')
plt.xlabel('Hayatta Kalma Durumu')
plt.ylabel('Yolcu Sayısı')

# Grafik 2: Bilet Sınıflarına Göre Hayatta Kalma Oranları
plt.subplot(1, 2, 2)
sns.barplot(x='Pclass', y='Survived', data=df_clean, palette='pastel', errorbar=None)
plt.title('Bilet Sınıfına Göre Kurtulma Oranı')
plt.xlabel('Bilet Sınıfı (1=En İyi, 3=En Ucuz)')
plt.ylabel('Hayatta Kalma Oranı')

plt.tight_layout()
plt.show()
```
